# Separating the Predictors and Target

```python
X = df[['Avg. Session Length','Time on App', 'Time on Website','Length of Membership']]  #X=Predictors(Feature)
X
```

```python
y=df['Yearly Amount Spent'] #y=target(label)
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

# Train the model

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train) # Basically fit means training the model with features --> "X_train" 
# Everytime X_train is passed it should return corresponding y_train
```

In machine learning, fit() is used to train the model by learning the relationship between the input features X and the target y. It adjusts the model’s parameters (like weights) to minimize errors and make accurate predictions.


```python
model.coef_  
# model.coef_ --> gives you the value of m (the slope).

# If you have multiple features, it gives an array of coefficients — one for each feature.
```

```python
# The coefficients in a dataframe
cdf = pd.DataFrame(model.coef_,X.columns,columns=['Coef'])
print(cdf)
```

# Create Predictions

```python
y_pred=model.predict(X_test)
y_pred
```

#### Graphical Evaluations of Prediction

```python
sns.scatterplot(x=y_pred, y=y_test)
plt.xlabel("Predictions")
plt.ylabel("Actual") #y_test's value which is yearly amount spent
plt.title("Predicted vs Actual")
plt.show()
```

```python
# Drawing a regression line in prediction vs actual scattered points

# Create a DataFrame with predicted and actual values
results = pd.DataFrame({
    'Predicted': y_pred,
    'Actual': y_test
})

# Use lmplot to show the regression line
sns.lmplot(x='Predicted', y='Actual', data=results, scatter_kws={'alpha':0.3}, line_kws={'color': 'red'})
```

## Analytical Evaluations of the Errors

#### Precision, Recall, Accuracy, and F1 = for classification

#### MAE, MSE, RMSE, R² = for regression

#### In linear regression, we typically don’t compute `F1-score, accuracy, precision, or recall` because
- These metrics are used for classification problems, where the goal is to predict categories or labels (e.g., spam vs not spam).
- But Linear Regression is for regression problems, where the goal is to predict continuous values (like price, temperature, salary, etc.).

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import math
```

```python
print('Mean Absolute Error:',mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:',mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:',math.sqrt(mean_squared_error(y_test, y_pred)))
print("R^2 Score:", r2_score(y_pred, y_test))
```
